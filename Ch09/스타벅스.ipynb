```python
import random
import requests as rq
def get_rndCod():
    charset = '0123456789ABCDEFGHIJKLMNOPQRSTUVWXYZ'
    return ''.join(random.sample(charset,10))

def getGugunList(rndCod):
    result = []
    for i in range(1,18):
        url = 'https://www.starbucks.co.kr/store/getGugunList.do?sido_cd='
        
        cd = '0'
        if i < 10:
            cd = '0' + str(i)
        else:
            cd = str(i)
        
        url = url +cd+'&rndCod='+rndCod
        
        res= rq.post(url).json()
        result.append(res['list'])
    return result
list = getGugunList(rndCod)
```
