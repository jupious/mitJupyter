```python
conn = cx_Oracle.connect('aaa/bbb@localhost:1521/xe')
cur = conn.cursor()

result = cur.execute("select * from todo")
all_data = result.fetchall()    #실행결과 한 행을 튜플로 전체 결과를 리스트에 담음
print(all_data)
```

```python
result = cur.execute("select column_name from user_tab_columns where table_name=UPPER('todo')") #전체 컬럼명 조회 (where문으로 특정 테이블 조회가능)
all_cols = result.fetchall()
all_cols = [r[0] for r in all_cols]
all_cols
```

```python
import pandas as pd
df = pd.DataFrame(all_data, columns= all_cols) 
df
```

```python
sql = "insert into todo(no, str) values(todo_key.nextval, '파이썬에서 테스트중입니다.')"
cur.execute(sql)
conn.commit() #실제 db에 커밋하기
```

```python
sql = "insert into todo(no, str) values(todo_key.nextval, :str)"
cur.execute(sql, str = '쿼리에 :을 통해 변수를 대입할수있음')
conn.commit()
```

```python
sql = "insert into todo values(todo_key.nextval, :todo_date, :str, :iscomplete)"    #:을 통해 변수를 할당할때는 테이블의 컬럼명과 같아야됨;;
cur.execute(sql,('2023-11-20','변수를 튜플로 삽입할수도 있다고함',0))
conn.commit()
```

```python
cur.close() #DB연결 종료
conn.close()
```
