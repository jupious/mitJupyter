```python
import urllib.request
from urllib.request import urlretrieve as rq
import json
import pandas as pd
from PIL import Image
from openpyxl import Workbook       #엑셀파일접근?용
from openpyxl.drawing.image import Image  as wImg#엑셀파일 이미지컨트롤?용
from io import BytesIO


client_id = "aSjJoPTQRKeEeSDU9u7r"
client_secret = "LWpn2MXoQA"
encText = urllib.parse.quote("노트북")
result_dict = {}
brand_list = []
lowcost = []
imgUrl = []
path = './images/'
thumbpath = './thumbnails/'
counter = 2

def cellWH(imgW, imgH):
  cellW = imgW * 63.2/504.19
  cellH = imgH * 225.35/298.96
  return (cellW, cellH)

wb = Workbook()
ws = wb.active
ws['A1'] = '브랜드'
ws['B1'] = '최저가'
ws['C1'] = '이미지'
col_width, row_height = cellWH(255, 200)
ws.column_dimensions['C'].width = col_width
for page in range(1,3):
  url = "https://openapi.naver.com/v1/search/shop.json?query=" + encText # JSON 결과
  url += "&display=20&start="+str(page)
  request = urllib.request.Request(url)
  request.add_header("X-Naver-Client-Id",client_id)
  request.add_header("X-Naver-Client-Secret",client_secret)
  response = urllib.request.urlopen(request)
  rescode = response.getcode()
  if(rescode==200):
      response_body = response.read().decode('utf-8')
      result = json.loads(response_body)
      #print(response_body)
      for items in result['items']:
          if items['brand'] == "":
            ws['A'+str(counter)] = "브랜드 없음"
          else:
            ws['A'+str(counter)] = items['brand']
          ws['B'+str(counter)] = items['lprice'] + "원"
          if items['image'] == "":
             pass
          else:
             imUrl = items['image']
             savename = imUrl.split("/")[-1]
             btImg = urllib.request.urlopen(imUrl).read()
             opImg = Image.open(BytesIO(btImg))
             thumb = opImg.convert('RGB').resize((255,200))
             thumb.save(thumbpath+savename)
             thumbnail = wImg(thumbpath+savename)
             ws.add_image(thumbnail, 'C'+str(counter))
             ws['C'+str(counter)] = ""
             ws.row_dimensions[counter].height = row_height
          counter += 1
      

  else:
      print("Error Code:" + rescode)
      break
wb.save("proto.xlsx")
wb.close()

```
