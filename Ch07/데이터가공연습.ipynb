```python
import statsmodels.api as sm
from statsmodels.formula.api import ols
accident = pd.read_csv('./통계.csv', encoding='cp949')
# acd = accident.columns.difference(['사고','연도'])
# acd_list = list(acd)
# sm.graphics.plot_partregress('사고','연도',acd_list, data=accident,ret_coords=True)
# #plt.xticks(accident['연도'])
# plt.show()

pr = ols('사고 ~ 연도',accident).fit()
print(pr.summary())

#test_df = pd.DataFrame({'연도':[2023]}) #꼭 데이터프레임으로 넣어줄 필요는 없음
pr.predict({'연도':[2023,2024]})
```
