```python
# 네이버 검색 API 예제 - 블로그 검색
# 4월03일 블로그에 쓴 글중 낙엽에 대한 제목을 다 수집하고 싶다. 
import urllib.request #http 요청
import json #json 데이터 처리
import pandas as pd #표 형태의 데이터를 처리하기 위한 모듈
import time
client_id = "aSjJoPTQRKeEeSDU9u7r"
client_secret = "LWpn2MXoQA"
encText = urllib.parse.quote("낙엽") #url 인코딩 변환
#주소만들고
title_list=[]
loop=True
for number in range(1,3):
    time.sleep(0.1) #1초 기다림
    url = "https://openapi.naver.com/v1/search/blog?query=" + encText # JSON 결과
    url = url+"&display=100&sort=date&start="+str(number)
    # url = "https://openapi.naver.com/v1/search/blog.xml?query=" + encText # XML 결과
    #헤더정보 추가
    request = urllib.request.Request(url)
    request.add_header("X-Naver-Client-Id",client_id)
    request.add_header("X-Naver-Client-Secret",client_secret)
    #요청
    response = urllib.request.urlopen(request)
    rescode = response.getcode()
    if(rescode==200):
        response_body = response.read().decode('utf-8') #읽어온데이터
        #print(response_body)
        json_data=json.loads(response_body) #json 데이터 읽기
        for i in json_data['items']:
            #print(i['title'])
            if i['postdate'] =='20231101':
                title_list.append(i['title'])
                print(i['title'])
            else:
                loop=False
                break   
    else:
        print("Error Code:" + rescode)
    if loop==False:
        break
#파일 저장하는법
df=pd.DataFrame(title_list)
df.to_csv("낙엽검색결과.csv",encoding="utf-8",index=False)
```
